```python
from tkinter import *
from tkinter import ttk
```

```python
main = Tk()
main.title("Calendar")
main. geometry()

label1 = Label(main, text="일")
label1.config(font=('굴림', 18))
label1.grid(row=0, column=0, columnspan=1)
label2 = Label(main, text="월")
label2.config(font=('굴림', 18))
label2.grid(row=0, column=1, columnspan=1)
label3 = Label(main, text="화")
label3.config(font=('굴림', 18))
label3.grid(row=0, column=2, columnspan=1)
label4 = Label(main, text="수")
label4.config(font=('굴림', 18))
label4.grid(row=0, column=3, columnspan=1)
label5 = Label(main, text="목")
label5.config(font=('굴림', 18))
label5.grid(row=0, column=4, columnspan=1)
label6 = Label(main, text="금")
label6.config(font=('굴림', 18))
label6.grid(row=0, column=5, columnspan=1)
label7 = Label(main, text="토")
label7.config(font=('굴림', 18))
label7.grid(row=0, column=6, columnspan=1)

boxes = []
str1 = []
for i in range(0,42):
    str1.append(StringVar())
    textbox = ttk.Entry(main, width=5, textvariable=str1[i])
    boxes.append(textbox)
    
k = 0    
for i in range(1,7):
    for j in range(0,7):
        boxes[k].grid(row=i, column=j, columnspan=1)
        boxes[k].insert("", k)
        boxes[k].config(justify="right")
        boxes[k].config(font=('굴림',18))
        k = k+1
        
btn1 = Button(main, text="이전달", width=7, command=lambda:prevMonth())
btn1.grid(row=7, column=0, columnspan=1)
btn2 = Button(main, text="다음달", width=7, command=lambda:nextMonth())
btn2.grid(row=7, column=6, columnspan=1)

strYear = StringVar()
textYear = ttk.Entry(main, width=5,textvariable=strYear)
textYear.config(justify='center')
textYear.config(font=('굴림', 18))
textYear.grid(row=7, column=1, columnspan=1)

strMonth = StringVar()
textMonth = ttk.Entry(main, width=5,textvariable=strMonth)
textMonth.config(justify='center')
textMonth.config(font=('굴림', 18))
textMonth.grid(row=7, column=3, columnspan=1)

label1 = Label(main, text="년")
label1.config(font=('굴림',18))
label1.grid(row=7, column=2, columnspan=1)
label2 = Label(main, text="월")
label2.config(font=('굴림',18))
label2.grid(row=7, column=4, columnspan=1)

btn3 = Button(main, text="확인", width=7, command=lambda:inputYM())
btn3.grid(row=7, column=5, columnspan=1)

import datetime
# textYear에 현재 날짜의 년도를 표시
textYear.insert(0, datetime.datetime.now().year)
# textMonth에 현재 날짜의 월을 표시
textMonth.insert(0, datetime.datetime.now().month)
# 현재날짜의 년과 월을 y, m 변수에 저장
y = datetime.datetime.now().year
m = datetime.datetime.now().month
# prnCalendar 함수호출( 전달인수 y,m)
prnCalendar(y,m)

main.mainloop()
```

```python
import calendar
def prnCalendar(y,m):
    # textbox 42개의 내용을 전부 삭제합니다
    for i in range(0,42):
        boxes[i].delete(0,"")
    # 전달받은 년과 월을 이용해서 해당월의 1일자 요일과, 마지막날짜를 얻습니다
    n,endDay = calendar.monthrange(y,m)
    print(n,endDay)
    # 월요일이 0이므로 n = n+1연산하고 0번째 건너뛰고 1번째 textbox부터 1일이 입력되게 합니다
    n = n+1
    if n==7:  # n=6이면 다음줄로 이동하므로 7일때 0으로 이동시
        n=0
    # n번째 textbox(boxes[n])부터 endDay까지 날짜를 표시합니다
    k=1
    for i in range(n, n+endDay):
        boxes[i].insert(0, str(k))
        k=k+1
        
def prevMonth():
    # 현재 날짜의 년과 월을 얻어냅니다
    y = int(strYear.get())
    m = int(strMonth.get())
    # 이전달을 계산합니다
    if m==1:
        m=12
        y=y-1
    else:
        m=m-1
    # 계산한 이전달의 년월을 textbox에 입력합니다
    textYear.delete(0,"")
    textYear.insert(0, str(y))
    textMonth.delete(0,"")
    textMonth.insert(0, str(m))
    # prnCalendar 함수를 호출합니다
    prnCalendar(y,m)
    
def nextMonth():
    # 현재 날짜의 년과 월을 얻어냅니다
    y = int(strYear.get())
    m = int(strMonth.get())
    # 다음달을 계산합니다
    if m==12:
        m=1
        y=y+1
    else:
        m=m+1
    # 계산한 다음달의 년월을 textbox에 입력합니다
    textYear.delete(0,"")
    textYear.insert(0, str(y))
    textMonth.delete(0,"")
    textMonth.insert(0, str(m))
    # prnCalendar 함수를 호출합니다
    prnCalendar(y,m)
    
def inputYM():
    y = int(strYear.get())
    m = int(strMonth.get())
    prnCalendar(y,m)
```
