```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
marathon_2015_2017 = pd.read_csv("marathon_2015_2017.csv")
marathon_2015_2017.head(2)
```

### 남여 참가자의 연령별 기록 비교

```python
# 남자 참가자와 여자 참가자를, 각각의 성별에 따라 두 개의 데이터프레임으로 따로 저장
MALE_runner =  marathon_2015_2017[marathon_2015_2017['M/F']=='M']
FEMALE_runner = marathon_2015_2017[marathon_2015_2017['M/F']=='F']

# 남자 참가자의 최종 기록들만 따로 변수에 저장
y_male = MALE_runner['Official Time']
# 여자 참가자의 기록들만 따로 변수에 저장
y_female = FEMALE_runner['Official Time']

# 남자 참가자의 나이를 x_male에 담습니다
x_male = MALE_runner['Age']
# 여자 참가자의 나이를 x_male에 담습니다
x_female = FEMALE_runner['Age']
plt.figure(figsize=(20,10))
plt.plot(x_male, y_male, '.',color='b', alpha=0.3, markersize=12)
plt.plot(x_female, y_female, '.',color='r', alpha=0.3, markersize=12)
# 's' : 표식옵션 - s는 사각형 모양의 표식
# 'b' : 파란색
# alpha : 투명도
plt.xlabel("Age", fontdict={'size':16, 'color':'white'})
plt.ylabel("Official Time(Second)",  fontdict={'size':16, 'color':'white'})
plt.title("Distribution by Running Time and Age", fontdict={'size':30, 'color':'white'})
plt.show()

# Marker의 종류
# '.' : point marker
# 'o' : circle marker
# 'v' : 아래쪽 삼각형
# '^' : 위쪽 삼각형
# '<', '>' : 왼쪽 오른쪽 방향 삼각형
# '+' : 플러스 기호
# 'D' : 다이아몬드 기호

# markersize : 표시 모양의 크기 조절
```
