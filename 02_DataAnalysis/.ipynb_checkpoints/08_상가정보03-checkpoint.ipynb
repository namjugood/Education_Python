### 소상공인시장진흥공단 상가업소정보로 스타벅스, 이디야 위치 분석

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.rc("font", family="Malgun Gothic")
plt.rc("axes", unicode_minus=False)
from IPython.display import set_matplotlib_formats
set_matplotlib_formats("retina")
```

```python
df = pd.read_csv("상가업소정보_201912_01.csv", sep="|")
n = df.isnull().sum()
not_use = n.sort_values(ascending=False).head(9)
not_use_col = not_use.index
df = df.drop(not_use_col, axis=1)
cols = df.columns
cols_code = cols[cols.str.contains("코드|번호")]
df = df.drop(cols_code, axis=1)
df = df.drop(['대지구분명','지번본번지','지번주소','도로명','건물본번지'], axis=1)
df_seoul = df[df["시도명"]=="서울특별시"]
df_seoul.head(2)
```

```python
df_file = pd.read_csv("seoul_open_store.csv")
df_file.head(2)
```

```python
# 스타벅스와 이디야를 대상으로한 분포도 조사

# 스타벅스와 이디야만 필터링하여 df_cafe변수에 저장합니다
# (영문, 한글, 다른철자로 쓰여질 수 있는 모든 상호명을 필터링합니다)
df_seoul["상호명_소문자"] = df_seoul["상호명"].str.lower()
df_cafe = df_seoul.loc[df_seoul["상호명_소문자"].str.contains(\
                 "스타벅스|starbucks|starbugs|이디야|이디아|ediya|idiya")]
# 브랜드명이라는 컬럼을 생성해서 상호명으로 구분한 해당 컬럼에 '스타벅스', '이디야'를 채워넣습니다
df_cafe.loc[df_cafe["상호명_소문자"].str.contains(\
                                    "스타벅스|starbucks|starbugs"), "브랜드명"]="스타벅스"
df_cafe["브랜드명"] = df_cafe["브랜드명"].fillna("이디야")

```

```python
df_cafe[df_cafe['상권업종대분류명']=='생활서비스']
```

```python
df_cafe[df_cafe['상권업종대분류명']=='학문/교육']
```

```python
df_cafe[df_cafe['상권업종대분류명']=='소매']
```

```python
df_cafe = df_cafe[~df_cafe["상권업종대분류명"]\
                  .isin(["학문/교육","소매","부동산","생활서비스"])]
```

```python
# 하나의 분류명만 가지고도 할 수 있으나 위의 과정은 간혹 들어가있을 정상값 체크를 위함
# 대분류명을 살펴보고 해당되지 않는 대분류명은 삭제합니다
df_cafe["상권업종대분류명"].value_counts()

df_cafe = df_cafe[df_cafe["상권업종대분류명"].isin(["음식"])]
#df_cafe = df_cafe[~df_cafe["상권업종대분류명"].isin(["학문/교육","소매","부동산","생활서비스"])]
df_cafe["상권업종대분류명"].value_counts()
```

```python
# seaborn의 countplot을 이용하여 브랜드별 매장 개수를 표기하고
# text를 이용하여 실제 매장 개수를 차트 위, 해당 위치에 표시하세요
brand_count = df_cafe["브랜드명"].value_counts()
graph = sns.countplot(data=df_cafe, x="브랜드명", order=["이디야","스타벅스"])

for i, val in enumerate(brand_count.index):
    graph.text(x=i, y=brand_count[i], s=brand_count[i])
# 표시되는 숫자와 그래프의 숫자가 각자 표시하는 방법이 교차되어 달라져 수치가 엇갈리게되므로
# 오름차순 내림차순을 order나 sort_value로 조절해야 함
```

```python
import numpy as np
brand_count = df_cafe["브랜드명"].value_counts().sort_values()
# order보다는 sort_values를 사용하는 것이 편함
g = sns.countplot(data=df_cafe, x="브랜드명")
g.set_yticks(np.arange(0, 700, 100)) # 눈금 늘리기(numpy가 있어야 함)
for i, val in enumerate(brand_count.index):
    g.text(x=i, y=brand_count[i]+20, s=brand_count[i])
    
```

```python
import folium
from folium.plugins import MarkerCluster
lat = df_cafe["위도"].mean()
long = df_cafe["경도"].mean()
m = folium.Map([lat, long], zoom_start=12, tiles="Stamen Toner")
marker_cluster = MarkerCluster().add_to(m)

for i in df_cafe.index:
    lat = df_cafe.loc[i, "위도"].mean()
    long = df_cafe.loc[i, "경도"].mean()
    title = df_cafe.loc[i,"브랜드명"] + "-" + df_cafe.loc[i,"도로명주소"]
    icon_color = 'red'
    if df_cafe.loc[i, "브랜드명"]=="스타벅스":
        icon_color='blue'
    folium.Marker(
        [lat, long],
        icon = folium.Icon(color=icon_color),
        tooltip=title).add_to(marker_cluster)
m
```

#### GeoJSON 파일로드
* 단순 데이터 파일을 정리할 때 사용하는 파일(DB와는 다름)
* 구 별로 매장수를 표현하기 위해 GeoJSON 파일 로드
    * 파일 출처 : (https://github.com/southkorea/seoul-maps)
    * 다운로드가 가능한 주소 https://drive.google.com/open?id=13j8-_XLdPe0pptsqu8-uyE-0Ym6V2jw5

```python
# 구별, 브랜드별 점포수를 groupby를 이용하여 계산하고
# df_cafe_vs 변수에 저장합니다
df_cafe_vs = df_cafe.groupby(["시군구명","브랜드명"])["상호명"].count()
df_cafe_vs.head()
```

```python
df_cafe_vs = df_cafe_vs.reset_index()
df_cafe_vs.head()
```

```python
geo_path = 'seoul_municipalities_geo_simple.json'

import json
geo_json = json.load(open(geo_path, encoding="utf-8"))
geo_json["features"][0]['properties']["name"]
```

```python
# df_cafe_vs변수에 담긴 값을 통해 스타벅스의 매장 수를 구합니다
# '브랜드명' 컬럼으로 스타벅스만 가져옵니다
df_star = df_cafe_vs[df_cafe_vs["브랜드명"]=="스타벅스"]
df_star.head()
```

```python
m = folium.Map( [lat, long])
```

```python
folium.Choropleth(
    geo_data=geo_json,
    name='choropleth',
    data=df_star,
    columns=['시군구명','상호명'],
    key_on="feature.properties.name",  #geo_json의 필드명
    fill_color='YlGn',  # 표현색상
    fill_opacity=0.71,   # 면색의 투명도
    line_opacity=0.2,   # 선색의 투명도
    legend_name='스타벅스 매장수'
).add_to(m)
m
```

```python
df_star = df_cafe_vs[df_cafe_vs["브랜드명"]=="이디야"]
m = folium.Map([lat, long])
folium.Choropleth(
    geo_data=geo_json,
    name='choropleth',
    data=df_star,
    columns=['시군구명','상호명'],
    key_on="feature.properties.name",  #geo_json의 필드명
    fill_color='Blues',  # 표현색상
    fill_opacity=0.71,   # 면색의 투명도
    line_opacity=0.2,   # 선색의 투명도
    legend_name='이디야 매장수'
).add_to(m)
m
```
