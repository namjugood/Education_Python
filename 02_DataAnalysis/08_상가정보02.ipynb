## 베스킨라빈스, 던킨도너츠 위치 분석

### 특정 상호만 가져오기

* 여기에서는 베스킨라빈스와 던킨도너츠 상호를 가져와서 실습합니다.
* 위에서 pandas의 str.contains를 활용해 봅니다
* 상호명에서 브랜드명을 추출합니다
* 대소문자가 섞여있을 수도 있기 때문에 대소문자를 변환해 줍니다.
* 오타방지를 위해 베스킨라빈스의 영문명은 baskinrobbins, 던킨도너츠는 dunkindonuts입니다

```python
# 문자열을 소문자로 변경하는 메소드를 시용합니다. 컬럼을 만듭니다
df_seoul["상호명_소문자"] = df_seoul["상호명"].str.lower()
```

```python
df_seoul.head(1)
```

```python
# 베스킨라빈스의 검색
df_seoul.loc[df_seoul["상호명_소문자"].str.contains(\
                      "배스킨라빈스|베스킨라빈스|baskinrobbins"),"상호명_소문자"].shape
```

```python
# 던킨도너츠의 검색
df_seoul.loc[df_seoul["상호명_소문자"].str.contains(\
                      "던킨|dunkin"),"상호명_소문자"].shape
```

```python
# 둘의 검색결과를 필터링 대상으로 하여 필터링하고 df_31에 저장합니다
df_31 = df_seoul.loc[df_seoul["상호명_소문자"].str.contains(\
                      "배스킨라빈스|베스킨라빈스|baskinrobbins|던킨|dunkin")]
df_31.shape
```

```python
df_31.head(3)
```

```python
# 상호명_소문자 중 배스킨라빈스 관련 상호명을 "배스킨라빈스"라는 값으로 
# 새로운 컬럼 "브랜드명"에 채웁니다
df_31.loc[df_31["상호명_소문자"].str.contains(\
                    "배스킨라빈스|베스킨라빈스|baskinrobbins"),"브랜드명"] = "배스킨라빈스"
```

```python
# 브랜드명의 결측치(빈칸)에 던킨도너츠를 채웁니다
df_31["브랜드명"] = df_31["브랜드명"].fillna("던킨도너츠")
```

```python
df_31[["상호명","브랜드명"]].head()
```

```python
# 데이터가 제대로 모아졌는지 확인합니다
# "상권업종대분류명"을 value.counts를 통해 빈도수를 계산합니다.
df_31["상권업종대분류명"].value_counts()
```

```python
# "상권업종대분류명"컬럼에서 isin기능을 사용해서 "소매", "생활서비스" 인
# 데어티만 검색해 봅니다
df_31[df_31["상권업종대분류명"].isin(["소매","생활서비스"])]
```

```python
# &:and  |:or  ~~:not  -  not 연산자를 이용해서 소매, 생활서비스가 아닌 것만 다시 필터링합니다.
df_31 = df_31[~df_31["상권업종대분류명"].isin(["소매","생활서비스"])]
```

```python
df_31.shape
```

### 시각화

* plot으로 차트그리기

```python
# 브랜드명으로 각 브랜드의 개수를 표현합니다
df_31["브랜드명"].value_counts().plot.barh()
```

```python
# 브랜드명의 value_counts 값을 비율로 표시하고 이를 차트에 표현합니다
df_31["브랜드명"].value_counts(normalize=True).plot.barh()
```

```python
# seaborn의  countplot을 이용한 막대그래프를 그립니다
brand_count = df_31["브랜드명"].value_counts()
g = sns.countplot(data=df_31, x="브랜드명")

# x, y좌표에 brand_count값(실제값)을 표시합니다
for i, val in enumerate(brand_count.index):
    g.text(x=i, y=brand_count[i]+20, s=brand_count[i])
```

```python
# 시군구명으로 빈도수를 세고 브랜드명으로 색상을 다르게 표하는 countplot을 그립니다
plt.figure(figsize=(15,4))
g = sns.countplot(data=df_31, x="시군구명", hue="브랜드명")
```

```python
# seaborn의 scatterplot으로 hue에 브랜드명을 지정해서 시각화합니다
sns.scatterplot(data=df_31, x="경도", y="위도", hue="브랜드명")
```

```python
# Lat : 37.495593  Long : 127.039032  tooltip : '던킨도너츠'의 값으로 
# forlium 맵에 마킹하세요
import folium
lat = df_31["위도"].mean()
long = df_31["경도"].mean()
m = folium.Map([lat, long], zoom_start=11)
folium.Marker([37.495593, 127.039032], tooltip='던킨도너츠').add_to(m) # 좌표 오기
m
```

* df_31의 위도값들과 경도값들을 이용하여 모두 마킹해주세요
* tooltip값은 브랜드명-도로명주소로 설정합니다
* zoom_start는 12 또는 11 중 적절히 설정하세요
* 던킨도너츠는 빨간색, 배스킨라빈스는 파란색으로 마킹하세요

```python
m = folium.Map([lat, long], zoom_start=12)
for i in df_31.index:
    lat = df_31.loc[i,"위도"]
    long = df_31.loc[i,"경도"]
    title = df_31.loc[i,"상호명"]+"-"+df_31.loc[i, "도로명주소"]
    icon_color="blue"
    if df_31.loc[i, '브랜드명']=="던킨도너츠":
        icon_color="red"
        
    folium.Marker(
        [lat, long],
        icon=folium.Icon(color=icon_color),
        tooltip=title).add_to(m)
m # 화면송출을 안해서 안보였음
```

## MarkerCluster로 표현하기

```python
from folium.plugins import MarkerCluster
m = folium.Map([lat, long], zoom_start=12)

marker_cluster = MarkerCluster().add_to(m)

for i in df_31.index:
    lat = df_31.loc[i,"위도"]
    long = df_31.loc[i,"경도"]
    title = df_31.loc[i,"상호명"]+"-"+df_31.loc[i, "도로명주소"]
    icon_color="blue"
    if df_31.loc[i, '브랜드명']=="던킨도너츠":
        icon_color="red"
    folium.Marker(
        [lat, long],
        icon=folium.Icon(color=icon_color),
        tooltip=title).add_to(marker_cluster)
m # 화면송출을 안해서 안보였음
```
