### 데이터 타입 변경
* 분양가격이 object(문자)타입으로 되어있습니다. 문자열 타입을 계산할 수 없기 때문에 수치 데이터로 변경해 줍니다. 결측치가 섞여있을 때 변환이 제대로 되지 않습니다. 그래서 pd.to_numeric을 통해 데이터의 타입을 변경합니다.

```python
df_last["분양가격"] = pd.to_numeric(df_last["분양가격(㎡)"], errors='coerce')
# errors='coerce'를 넣어야 에러가 나지 않음 => 수치를 NaN값으로 변경
df_last["분양가격"].head()
```

```python
df_last.info()
```

```python
type(pd.np.nan)
```

### 평당 분양가격 구하기
* 현재 사용중인 2013년부터의 데이터는 평당분양가격 기준으로 되어있습니다.
* 분양가격을 평당기준으로 보기위해 3.3을 곱해서 "평당분양가격"컬럼을 만들어 추가해 줍니다

```python
df_last["평당분양가격"] = df_last["분양가격"]*3.3
df_last.head()
```

```python
df_last.info()
```

### 규모구분을 전용면적 컬럼으로 변경
규모구분 컬럼은 전용면적에 대한 내용이 있습니다. 전용면적이라는 문구가 공통적으로 들어가고 규모구분보다는 전용면적이 좀 더 직관적이기 때문에 전용면적이라는 컬럼을 새로 만들어주고 기존 규모구분의 값에서 전용면적, 초과, 이하 등의 문구를 빼고 간결하게 만들어봅니다.

이때 str의 replace 기능을 사용해서 예를 들면 "전용면적 60㎥ 초과 85㎥ 이하"라면 60㎥~85㎥로 변경해 줍니다

```python
# "전용면적" 삭제
df_last["전용면적"] = df_last["규모구분"].str.replace("전용면적", "")
# "초과"를 "~"로 대체
df_last["전용면적"] = df_last["전용면적"].str.replace("초과", "~")
# "이하" 삭제
df_last["전용면적"] = df_last["전용면적"].str.replace("이하", "")
# 스페이스바 하나에 해당하는 공백을 ""로 대체(삭제)
df_last["전용면적"] = df_last["전용면적"].str.replace("", "")
df_last["전용면적"]
```

```python
# 필요없는 컬럼을 삭제합니다
df_last = df_last.drop(["규모구분","분양가격(㎡)"], axis=1)
df_last.head(5)
```

```python
df_last.info()
```

### groupby로 데이터 집계하기
groupby를 통해 데이터를 그룹화해서 연산을 해봅시다.

```python
df_last.groupby(["지역명"])["평당분양가격"].mean()
```

```python
# 전용면적으로 분양가격의 평균 계산
df_last.groupby(["전용면적"])["평당분양가격"].mean()
```

```python
# 전용면적, 지역명 별 평당 분양가격의 평균
df_last.groupby(["전용면적","지역명"])["평당분양가격"].mean()
```

```python
df_last.groupby(["전용면적","지역명"])["평당분양가격"].mean().unstack().round()
```

```python
g = df_last.groupby(["연도","지역명"])["평당분양가격"].mean()
g.unstack().transpose().round()
```

### Pivot table로 데이터 집계하기
* groupby로 했던 작업을 pivot_table로 똑같이 해봅니다.

```python
# 지역명 별 평균 평당 분양가격
pd.pivot_table(df_last, index=["지역명"], values=["평당분양가격"],\
              aggfunc="mean").round()
# index : 그룹화할 항목의 컬럼 이름
# values : 해당 함수로 집계 또는 계산할 수치형 데이터 컬럼 이름
# aggfunc : 사용할 함수의 이름
```

```python
# 연도, 지역명 평당 분양 가격
p = pd.pivot_table(df_last, index=["연도", "지역명"], values="평당분양가격")
# pivot_table()함수는 구성속성 중 aggfunc가 기본값으로 mean이 지정되어 있습니다.
# 대상 values 컬럼이 숫자데이터일 때 적용, 문자라면 count
p.unstack().transpose().round()
```

### Unstack()의 반대방향 변형함수 Melt()
* pandas의 melt를 이용하여 unstack() 반대방향 변형을 할 수 있습ㄴ디ㅏ. 각 컬럼의 이름으로 존재하는 데이터들을 행으로 변형하여 조합 또는 어울림의 관계로 배치합니다.

```python
# pd.melt를 사용하여, df_first_melt 변수에 담습니다
df_first_melt = df_first.melt(id_vars="지역", var_name="기간", \
                             value_name="평당분양가격")
df_first_melt.head(10)
```

```python
df_last.head(1)
```

```python
df_first_melt.columns = ["지역명", "기간", "평당분양가격"]
df_first_melt.head(1)
```

```python
date = "2013년12월"
# split을 통해 "년"을 기준으로 텍스트를 분리해 봅니다
date.split("년")
```

```python
# 리스트의 인덱싱을 사용하여 연도만 추출
int(date.split("년")[0])
```

```python
# 리스트의 인덱싱과 replace를 이용하여 월을 제거합니다
int(date.split("년")[-1].replace("월",""))
```

```python
# 선택 1
def parse_year(date):
    year = int(date.split("년")[0])
    return year
df_first_melt["연도"] = df_first_melt["기간"].apply(parse_year)
```

```python
# 선택2
df_first_melt["연도"] = df_first_melt["기간"].apply(\
                                               lambda x : int(x.split("년")[0]))
```

```python
df_first_melt["월"] = df_first_melt["기간"].apply(\
                                   lambda x : int(x.split("년")[-1].replace("월","")))
```

```python
df_first_melt.tail(3)
```

```python
df_last.head(3)
```

```python
df_last_prepare = df_last.loc[df_last["전용면적"]=="전체", \
                             ["지역명","연도","월","평당분양가격"]]
df_last_prepare.head(1)
```

```python
df_first_prepare = df_first_melt[["지역명","연도","월","평당분양가격"]]
df_first_prepare.head(1)
```

```python
df = pd.concat([df_first_prepare, df_last_prepare])
df.shape
```

```python
df["연도"].value_counts(sort=False)  # 값에 의한 정렬없이 결과 출력
```

```python
t = pd.pivot_table(df, index="연도", columns="지역명", \
                  values="평당분양가격").round()
t.T  #t.transpose()
```

```python
# 히트맵 작성
plt.figure(figsize=(15,7))
sns.heatmap(t.T, cmap="Blues", annot=True, fmt=".0f")
# cmap : 색깔변경
# annot : 숫자표기
# fmt : 소수점 관리
```

```python
# barplot : 연도별, 지역명 별 평당 분양가격
plt.figure(figsize=(12,4))
sns.barplot(data=df, x="연도", y="평당분양가격", hue="지역명", ci=None)
plt.legend(bbox_to_anchor=(1.02, 1), loc=2)
```
